# person-300 dataset overview

person-300 dataset is subset of PASCAL VOC by Martin   
It contains `Annotations`, `JPEGImages`, `voc.names`

Folder `Annotations` has annotations in forms of VOC style xml

# 3. Verifying of Annotations

First of all, We have to verify where Annotations are right  
We're going to check a few lists  
1. image size 
2. bounding box of object

# 3.1. Verifying of Annotations - size of images

Previous, We saw xml   
`width : 500`  
`height : 375`  
`depth : 3`  

```python
print(src.shape) #shape of a image is of `(height, width, depth)`
print("width  : ", src.shape[1])
print("height : ", src.shape[0])
print("depth  : ", src.shape[2])
```

# 3.1. Verifying of Annotations - bounding box of objects

```python
import cv2
bbox_obj1 = (293, 162, 419, 375)
bbox_obj2 = (114, 165, 228, 373)
bbox_obj3 = (5, 172, 116, 373)

red = (255, 0, 0)
green = (0, 255, 0)
blue = (0, 0, 255)
```

```python
dst = src.copy()
dst = cv2.rectangle(dst,(bbox_obj1[0:2]),(bbox_obj1[2:4]), red, 5)
dst = cv2.rectangle(dst,(bbox_obj2[0:2]),(bbox_obj2[2:4]), green, 5)
dst = cv2.rectangle(dst,(bbox_obj3[0:2]),(bbox_obj3[2:4]), blue, 5)
plt.imshow(dst)
```
